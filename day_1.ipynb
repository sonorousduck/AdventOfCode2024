```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("puzzle_1.txt", delimiter="   ", engine="python", header=None)
data_numpy = data.to_numpy()
print(data_numpy)

# data_numpy = np.array([[3, 4],
#                        [4, 3],
#                        [2, 5],
#                        [1, 3],
#                        [3, 9],
#                        [3, 3]])
```

```python
column_1 = data_numpy[:, 0]
column_2 = data_numpy[:, 1]
column_1.sort()
column_2.sort()
print(column_1)
```

```python
total_difference = 0
for i in range(len(column_1)):
    # print(abs(column_1[i] - column_2[i]))
    total_difference += abs(column_1[i] - column_2[i])
print(total_difference)
```

```python
# Computing Similarity Score now

# If we have the sorted list, then we can just compare until there aren't anymore, then increase the left side until it matches the right

col_1_dict = dict.fromkeys(column_1, 0)

for i in range(len(column_2)):
    if column_2[i] in col_1_dict:
        col_1_dict[column_2[i]] += 1


total = 0

for i in column_1:
    total += col_1_dict[i] * i
print(total)
```
